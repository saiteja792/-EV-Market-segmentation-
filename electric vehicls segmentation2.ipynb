```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = 'C://Users//saite//OneDrive//Desktop//Indian automoble buying behavour study 1.0.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Demographic Segmentation: Profession Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Profession')
plt.title('Profession Distribution of Car Buyers')
plt.xlabel('Profession')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
# Behavioral Segmentation: Loan Status
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Personal loan', hue='House Loan')
plt.title('Loan Status of Car Buyers')
plt.xlabel('Personal Loan')
plt.ylabel('Count')
plt.legend(title='House Loan')
plt.grid(True)
plt.show()

```

```python
# Behavioral Segmentation: Car Make vs. Price
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Make', y='Price')
plt.title('Car Make vs. Price')
plt.xlabel('Car Make')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
