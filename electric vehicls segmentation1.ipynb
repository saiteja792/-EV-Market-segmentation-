```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset

df = pd.read_csv('C://Users//saite//OneDrive//Desktop//ElectricCarData_Clean.csv')



# Scatter plot for Price vs Range
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PriceEuro', y='Range_Km', hue='Brand', palette='viridis')
plt.title('Market Segmentation by Price and Range')
plt.xlabel('Price (Euro)')
plt.ylabel('Range (Km)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()


```

```python
# Box plot for Efficiency by Brand
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Brand', y='Efficiency_WhKm')
plt.title('Efficiency Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Efficiency (Wh/Km)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Brand', y='TopSpeed_KmH')
plt.title('Top Speed Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Top Speed (Km/H)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PriceEuro', y='AccelSec', hue='Brand', palette='coolwarm')
plt.title('Acceleration vs. Price')
plt.xlabel('Price (Euro)')
plt.ylabel('Acceleration (0-100 km/h in seconds)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Efficiency_WhKm', y='Range_Km', hue='Brand', palette='plasma')
plt.title('Range vs. Efficiency')
plt.xlabel('Efficiency (Wh/Km)')
plt.ylabel('Range (Km)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Brand', hue='RapidCharge')
plt.title('Rapid Charging Capability by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['PowerTrain'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Powertrain Distribution')
plt.ylabel('')
plt.show()
```
