```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
charging_data = pd.read_csv('C://Users//saite//OneDrive//Desktop//ev-charging-stations-india.csv')
charging_data.head()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(charging_data.pivot_table(index='type', columns='city', values='name', aggfunc='count'), cmap='coolwarm')
plt.title('Heatmap of Charging Stations')
plt.show()
```

```python
# Psychographic Segmentation Visualization
plt.figure(figsize=(10, 6))
charging_data['state'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('city and EV Preferences')
plt.ylabel('')
plt.show()
```
